```python
# Write a program to print the first and last character of a given string.
string = ("sai kiran naidu")
first_word = string[0]
last_word = string[-1]
print(first_word)
print(last_word)
```

```python
# Print the first three characters of a given string
string = ("hello good morning")
print(string[0:3])
```

```python
# Count how many times the letter 'o' appears in a given string
string = " the letter of in a called optimization o"
print(string.count("o"))
```

```python
# Remove all occurrences of "e" from a string.
string = "remove all occurences   of a string e"
print(string.replace("e", " "))
```

```python
# Write a program to remove all spaces from a given string.
string = ("sai kiran naidu")
remove_spaces  = string.replace(" ","")
print(remove_spaces)
```

```python
 # Write a program to check if a string contains only numbers.
string = input("Enter the string: ")
if string.isdigit():
    print("The string contains only numbers")
else:
    print("The string does not contain only numbers")
```

```python
# Write a program to convert a string to uppercase and lowercase.
string = ("GANDHAM")
upper_string =string.upper()
lower_string = string.lower()
print(upper_string)
print(lower_string)
```

```python
# Take two strings as input and merge them.
string1 = input("enter the first string: ")
string2 = input("enter the second string: ")
print(string1 + " " + string2)
```

```python
# Check if a username contains only letters and numbers
string = "username123"
if string.isalnum():
    print("The username contains only letters and numbers.")
else:
    print("The username contains only other characters.")

```
