```
#Create a dictionary with name, age, and course and print it
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
print(student)
```

```
#Print all keys of a dictionary.
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
print(student.keys())
```

```
#Print all values of a dictionary
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
print(student.values())
```

```
#Add a new key-value pair to a dictionary
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
student["grade"] = "A"
print(student)
```

```
#Update a value in a dictionary
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
student["age"] = 22
print(student)
```

```
#Write a program to remove an item using pop()
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
student.pop("course")
print(student)
```

```
#Write a program to delete a key using del
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
del student["age"]
print(student)
```

```
#Write a program to merge two dictionaries
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
student1 = {
    "grade": "A"
}
student.update(student1)
print(student)
```

```
#Write a program to remove the last inserted item using popitem()
student = {
    "name": "Sai",
    "age": 21,
    "course": "Python"
}
student.popitem()
print(student)
```
