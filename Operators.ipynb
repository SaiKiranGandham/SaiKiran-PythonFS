**Arithmetic Operators**

```python
a = 10
b = 3

print(a + b)
print( a - b)
print( a * b)
print( a / b)
print(a % b)
print( a ** b)
print( a // b)
```

**Assignment Operators**

```python
x = 10
x +=5
print(x)
x-=6
print(x)
x*=5
print(x)
```

**Comparison Operators**

```python
a = 10
b = 20
print(a==b)
print(a!=b)
print(a<b)
print(a>b)
print(a>=b)
print(a<=b)
```

**Logical Operators**

```python
x = True
y = False
print(x and y)
print(x or y)
print(not a)
```

**Bitwise Operators**

```python
x = 10
y = 30
print(x ^ y)
print(x | y)
print(x & y)
print(x >> y)
print(x << y)
```

**Identity Operators**

```python
x = [1, 2, 3]
y = x
z = [1, 2, 3]
print(x is y)
print(x is z)
print(x is not z)
```

**Membership Operators**

```python
numbers = [10, 20, 30, 40]

print(20 in numbers)
print(50 in numbers)
print(50 not in numbers)
```
